```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None) # display all columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

```python
confirmedStays = pd.read_csv('/Users/saman.tamkeen/Downloads/export (7).csv',dtype={'voyagercityid':'str','hotel':'str'})
confirmedStays.head()
```

```python
def normalize(df):
    df['confirmedbookings_log2'] = np.log2(confirmedStays['confirmedbookings']+1)
    desc = df['confirmedbookings_log2'].describe()
    q1 = desc['25%']
    q2 = desc['75%']
    iqr = q2 - q1
    _max = min(q2 + 1.5*iqr, desc['max'])
    df.loc[df['confirmedbookings_log2'] > _max, 'confirmedbookings_log2'] = _max
    df['confirmedbookings_norm'] = df['confirmedbookings_log2'] / _max
    return df
```

```python
confirmedStays.groupby('voyagercityid').apply(normalize)
```
